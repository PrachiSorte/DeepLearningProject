**Resizing the image using Pillow library**

```python
#Before feeding to NN, all images must be in common dimension
from PIL import Image
```

```python
#resize image
img=Image.open('/content/dog.jpg')
img_resized=img.resize((200,200))
```

```python
#save the resixed image
img_resized.save('/content/dog_image_resized.jpg')
```

```python
#displaying image from numpy array (to see dimensions of image)
img_res=mpimg.imread('/content/dog_image_resized.jpg')
img_res_plot=plt.imshow(img_res)
plt.show()
```

```python
#to check shape or size of image
print(img_res.shape)
```

**Convert RGB images to Gray Scale Image using OpenCV**

```python
#Importing opencv library (CV: Computer Vision)
import cv2
```

```python
img=cv2.imread('/content/dog.jpg')
```

```python
type(img)
```

```python
img.shape
```

```python
grayscale_img=cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
```

```python
type(grayscale_img)
```

```python
grayscale_img.shape  #(so channel 3 is missing as there is only 1 channel for grayscale image)
```

```python
#to display image from numpy array
#cv2.imshow() will display image(spyder, pycharm) but not allowed in colab

from google.colab.patches import cv2_imshow
```

```python
#displaying grayscale image
cv2_imshow(grayscale_img)
```

```python
# To save grayscale image using opencv
cv2.imwrite('/content/grayscale_dog_image.jpg',grayscale_img)
```



* Converting an color image into gray scale image saves a
  lot of space and computation time
* only applicable use cases where color of image is not
