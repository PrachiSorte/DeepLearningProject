```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

Data Collection and Preprocessing

```python
#loading the data from sklearn
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
#form of dictionary where key is data and values
# WHen we give all features it tells us whether a particular tumor is benign (does not move to other parts of body) or malignant (cancer tumour--spread)
breast_cancer_dataset
```

```python
# Load the data into dataframe

data_frame=pd.DataFrame(breast_cancer_dataset.data,columns=breast_cancer_dataset.feature_names)
```

```python
data_frame.head()
```

```python
# Adding the target column to the data frame
data_frame['target']= breast_cancer_dataset.target
```

```python
data_frame.tail()
```

```python
# to find out number of rows and columns in the dataset
data_frame.shape
```

```python
# To get some info about data
data_frame.info()
```

```python
#check missing values
data_frame.isnull().sum()
```

```python
#Statistical measures about the data
data_frame.describe()
```

```python
#Checking the distribution of target variable to check whether its a balance or imbalance dataset
# Here 1 --> Benign and 0 --> Malignant  (we can assume its a balace dataset as there is no hufe difference)
data_frame['target'].value_counts()
```

By doing this we can infer that target 0( malignant) have high values in all features compared to target 1 (Benign) features

```python
#grouping all data points with label 0 and finding their mean(avg) of all features and same with label 1 data points
data_frame.groupby('target').mean()
```

```python
#Split dataframe into features and target
X=data_frame.drop(columns='target',axis=1)
Y=data_frame['target']
```

```python
X
```

```python
Y
```

```python
#Split the data into train (80%) and test dataset (20%)
X_train, X_test, Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
print(X_train.shape,X_test.shape,Y_train.shape,Y_test.shape)
```

**Standardize the data**

```python
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()
X_train_std=scaler.fit_transform(X_train)
X_test_std=scaler.transform(X_test)   # never use fit_transform with test data
```

```python
#no need to apply scaler to Y values as its already having values 0 and 1
X_train_std # all values are in the same range
```

Building Neural Network (input-hidden-output layer)

```python
#imprort tensorflow and keras
import tensorflow as tf
from tensorflow import keras
tf.random.set_seed(3)   # To get same ouput each time we run the code
```

```python
# Setting up the layers of neural network
# Flatten will convert data into a single dimensional array as we have to pass all features at once as array
model=keras.Sequential([
    keras.layers.Flatten(input_shape=(30,)),    #first layer-Input layer - contains 30 input features
    keras.layers.Dense(20,activation='relu'),  #Hidden layer - contains hidden neurons and activation function
    keras.layers.Dense(2,activation='sigmoid')   #Output layer - contains 2 neurons as two classes are there and sigmoid activation function
])
```

```python
#compiling the neural network
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# Training the neural network:
# X_train-> no.of features, Y_train--> target or label for corresponsing input features
# validation_split is similar to train_test_split while model is training it do testing itself and
#epochs --> no. of iterations, how many times model is going thru data
# history -->to plot loss values and accuracy values while training going on
history=model.fit(X_train,Y_train,validation_split=0.1,epochs=10)
```

```python
# Training the neural network with standardized values ans see accuracy increases
history=model.fit(X_train_std,Y_train,validation_split=0.1,epochs=10)
```

**Visualizing accuracy and loss**

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')

plt.legend(['training data','validation data'],loc='lower right')
plt.show()
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')

plt.legend(['training data','validation data'],loc='upper right')
plt.show()
```

**Testing accuracy of model on test data**

```python
loss,accuracy=model.evaluate(X_test_std,Y_test)
print(accuracy)
```
